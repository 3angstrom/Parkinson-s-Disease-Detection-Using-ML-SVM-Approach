Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data Collection and Analysis

```python
# Loading the data form CSV file to a Pandas DataFrame

parkinsons_data = pd.read_csv('/content/parkinsons.csv')
```

Data Pre-Processing

```python
X = parkinsons_data.drop(columns=['name', 'status'], axis =1)
Y = parkinsons_data['status']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data to Training data & Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Data Standardization

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)

X_test = scaler.transform(X_test)
```

```python
print(X_train)
```

Model Training Machine Learning

Support Vector Machine Model

```python
model = svm.SVC(kernel='linear')
```

```python
#training the SVM model with training data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy score on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
print('Accuracy score of training data: ', training_data_accuracy)
```

```python
# accuracy score on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('Accuracy score of training data: ', test_data_accuracy)
```

Building a Predictive System

```python
input_data = (119.992,157.302,74.997,0.00784,0.00007,0.0037,0.00554,0.01109,0.04374,0.426,0.02182,0.0313,0.02971,0.06545,0.02211,21.033,0.414783,0.815285,-4.813031,0.266482,2.301442,0.284654)

# changing input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array
input_data_reshape = input_data_as_numpy_array.reshape(1, -1)

# standardize the data
std_data = scaler.transform(input_data_reshape)

# prediction
prediction = model.predict(std_data)
print(prediction)

if (prediction[0] == 0):
  print("The Person does not have Parkinsons Disease")

else:
  print("Person has Parkinsons")
```
